```python
import pandas as pd


df = pd.read_csv('household_power_consumption.txt', sep=';',
                 parse_dates={'Datetime': ['Date', 'Time']},
                 infer_datetime_format=True,
                 na_values=['?'],
                 low_memory=False)


print(df.info())
print(df.head())
print(df.describe())


missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)

df.fillna(method='ffill', inplace=True)
df.set_index('Datetime', inplace=True)


daily_df = df.resample('D').mean()

print(daily_df.shape)
print(daily_df.head())

from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

result = seasonal_decompose(daily_df['Global_active_power'], model='additive', period=365)
result.plot()
plt.show()



```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
import matplotlib.pyplot as plt


data = daily_df[['Global_active_power']].values
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(data)

def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i:i+seq_length])
        y.append(data[i+seq_length])
    return np.array(X), np.array(y)

sequence_length = 30  # one month of daily values
X, y = create_sequences(scaled_data, sequence_length)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]


model = Sequential([
    LSTM(50, return_sequences=False, input_shape=(X_train.shape[1], X_train.shape[2])),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(X_train, y_train, epochs=20, batch_size=16, validation_split=0.1)

y_pred = model.predict(X_test)
y_test_inv = scaler.inverse_transform(y_test)
y_pred_inv = scaler.inverse_transform(y_pred)

# Evaluation  of the metrics
mse = mean_squared_error(y_test_inv, y_pred_inv)
rmse = math.sqrt(mse)
mae = mean_absolute_error(y_test_inv, y_pred_inv)

print(f"MAE: {mae:.3f}")
print(f"RMSE: {rmse:.3f}")

#Actual vs predicted
plt.figure(figsize=(12,6))
plt.plot(y_test_inv, label='Actual')
plt.plot(y_pred_inv, label='Predicted')
plt.title('LSTM Forecasting')
plt.xlabel('Days')
plt.ylabel('Global Active Power (kW)')
plt.legend()
plt.show()


#ARIMA
ts_data = daily_df['Global_active_power'].dropna()

from statsmodels.tsa.stattools import adfuller

result = adfuller(ts_data)
print('ADF Statistic:', result[0])
print('p-value:', result[1])

#plots for the ARIMA parameters
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

ts_diff = ts_data.diff().dropna()

plt.figure(figsize=(12,6))
plt.subplot(121)
plot_acf(ts_diff, ax=plt.gca(), lags=40)
plt.subplot(122)
plot_pacf(ts_diff, ax=plt.gca(), lags=40)
plt.tight_layout()
plt.show()

from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(ts_data, order=(5, 1, 2))
model_fit = model.fit()
print(model_fit.summary())

# Forecasting future values
n_forecast = len(y_test)
forecast = model_fit.forecast(steps=n_forecast)

mse_arima = mean_squared_error(ts_data[-n_forecast:], forecast)
rmse_arima = math.sqrt(mse_arima)
mae_arima = mean_absolute_error(ts_data[-n_forecast:], forecast)

print(f"ARIMA MAE: {mae_arima:.3f}")
print(f"ARIMA RMSE: {rmse_arima:.3f}")

#Plots for ARIMA Forecast

plt.figure(figsize=(12,6))
plt.plot(ts_data[-n_forecast:].values, label='Actual')
plt.plot(forecast.values, label='Forecast')
plt.title('ARIMA Forecasting')
plt.xlabel('Days')
plt.ylabel('Global Active Power (kW)')
plt.legend()
plt.show()








```
